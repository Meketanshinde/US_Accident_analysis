```python
a=cities_by_accidents[cities_by_accidents == 1].count()
b=cities_by_accidents[cities_by_accidents == 2].count()
c=cities_by_accidents[cities_by_accidents == 35].count()
print(a,b,c)
```

* more than 1100 cities only reported 1 accident in 4 years,  

```python
high_accidents_cities = cities_by_accidents[cities_by_accidents >= 1000]
low_accidents_cities = cities_by_accidents[cities_by_accidents < 1000]
```

```python
(len(high_accidents_cities) / len(cities))*100
```

```python
sns.distplot(high_accidents_cities)
```

```python
sns.distplot(low_accidents_cities)
```

```python
under_1000 = cities_by_accidents[cities_by_accidents <= 1000]
accidents_1000_5000 = cities_by_accidents[(cities_by_accidents > 1000) & (cities_by_accidents <= 5000)]
# accidents_5000_100000 = cities_by_accidents[(cities_by_accidents > 5000) & (cities_by_accidents <= 100000)]
# accidents_10000_20000 = cities_by_accidents[(cities_by_accidents > 10000) & (cities_by_accidents <= 20000)]
# accidents_20000_40000 = cities_by_accidents[(cities_by_accidents > 20000) & (cities_by_accidents <= 40000)]
# accidents_40000_100000 = cities_by_accidents[(cities_by_accidents > 40000) & (cities_by_accidents <= 100000)]
# accidents_5000_10000.count(), accidents_10000_20000.count(), accidents_20000_40000.count(), accidents_40000_100000.count(),
# '>5000 & <=10000', '>10000 & <=20000', '>20000 & <=40000', '>40000 & <=100000',
accidents_more_5000 = cities_by_accidents[cities_by_accidents > 5000]

under_1000.count()
```

```python
plt.figure(figsize=(20,8))
x = np.array([under_1000.count(), accidents_1000_5000.count(), accidents_more_5000.count()])
mylabels = ['<=1000','>1000 & <=5000',  '>5000']
plt.title('Total cities  with Percentages as accidents intensity', fontsize = 20)
color = sns.color_palette("colorblind")
plt.pie(x, labels= mylabels,colors = color, autopct ='%0.1f%%')
plt.legend()
plt.show()
```

* Here we can clearly see that more than 95% of cities has less than 1000 of accidents
* And less than 5% of cities has more than 1000 accidents 

## Start Time 

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins= 24, kde=False)
```

* A major accidents happend between 3 to 5 pm, but if we see the graph the accidents started increasing from 12 am  and goes sideways or decreased after 3pm, lets see the tempraure at that time, or reason could be traffic. 
* Next Highest percentage is 6 am to 8 am (Probably because of the office time)

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins= 7, kde=False)
```

Is the distribution of accidents by hour is same on weekends as on week days ?

```python
SundayStartTime= df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(SundayStartTime.dt.hour, bins= 24, kde=False)
```

```python
SaturdayStartTime= df.Start_Time[df.Start_Time.dt.dayofweek == 5]
sns.distplot(SaturdayStartTime.dt.hour, bins= 24, kde=False)
```

On week days peak occures between 6 to 8 am where on weekends peak occures between 3 to 5 pm
